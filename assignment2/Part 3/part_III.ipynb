```python
import numpy as np
import matplotlib.pyplot as plt

data_loaded = np.loadtxt('output_t=5.txt', skiprows=1)

train_acc = data_loaded[:, 0]
train_loss = data_loaded[:, 1]
test_acc = data_loaded[:, 2]
test_loss = data_loaded[:, 3]

steps = np.arange(len(train_acc)) * 10

fig, axs = plt.subplots(1, 2, figsize=(12, 5))  # 2 Rows, 1 Column

# Plot training and testing accuracy
axs[0].plot(steps, train_acc, label='Train Accuracy')
axs[0].plot(steps, test_acc, label='Test Accuracy')
axs[0].set_title('Training and Testing Accuracy')
axs[0].set_xlabel('Steps')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# Plot training and testing loss
axs[1].plot(steps, train_loss, label='Train Loss')
axs[1].plot(steps, test_loss, label='Test Loss')
axs[1].set_title('Training and Testing Loss')
axs[1].set_xlabel('Steps')
axs[1].set_ylabel('Loss')
axs[1].legend()

fig.text(0.5, 0.99, 'T = 5', ha='center', va='top', fontsize=14)

# Display the plots
plt.tight_layout()
plt.show()
```
