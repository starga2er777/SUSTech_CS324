```python
import train_mlp_numpy

# For convenience, the train process is invoked here rather than using commands

# Default constants
DNN_HIDDEN_UNITS_DEFAULT = [20]
LEARNING_RATE_DEFAULT = 1e-2
MAX_EPOCHS_DEFAULT = 1500 # adjust if you use batch or not
EVAL_FREQ_DEFAULT = 10
BATCH_SIZE_DEFAULT = 800

train_acc, train_loss, test_acc, test_loss = train_mlp_numpy.train(x_train, y_train, x_test, y_test, DNN_HIDDEN_UNITS_DEFAULT, 
                                                                   LEARNING_RATE_DEFAULT, MAX_EPOCHS_DEFAULT, EVAL_FREQ_DEFAULT, BATCH_SIZE_DEFAULT)
```
